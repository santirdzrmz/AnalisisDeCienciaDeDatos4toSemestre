**¿Por qué es importante el modelado de datos en la fase de Integración? ¿Qué es un Modelo Entidad-Relación?** 

Es importante el modelado de datos en la fase de Integracion para que los datos esten organizados y sean precisos, al igual que es importante tener un panorama claro para tener consistencia y facilidad de consulta al crear las bases de datos. Un modelo Entidad Relacion es una representacion visual de datos, en donde se definen entidades, atributos, relaciones, cardinalidades y llaves. 

**¿El Modelo Entidad-Relación es un modelo Conceptual, Lógico o Físico?**

Es un modelo conceptual porque se quiere visualizar como se relacionan y estructuran los datos. 


**¿Cuáles son los elementos principales en un modelo Entidad-Relación?**

Son las entidades: cosas u objetos, por ejemplo, clientes, ubicaciones, productos etc. Los atributos: propiedades o variables, por ejemplo nombres edad, precio, etc.
Las relaciones: la edad de un cliente, el nombre de un cliente, etc. La cardinalidad, que explica cuantas ocurrencias de una entidad se relacionan con otra; 1:1, 1:N N:M. 
Y finalmente las claves primarias y foraneas.

**Indica un ejemplo de un modelo de datos lógico.**
Un modelo de datos logico seria un modelo relacional, en donde todo se relaciona exclusivamente a traves de las llaves foraneas. 


**¿Cuál es la diferencia entre modelo ER y modelo dimensional? Cuándo se utiliza cada uno?**
Un modelo ER normalmente se utiliza para sistemas transaccionales, donde es esencial mantener la eficiencia de las transacciones y sus relaciones (OLTP). Mientras que un modelo dimensional se usa para sistemas de analasis donde es mas importante la facilidad de uso y consulta, al igual que la escalabilidad del modelo (OLAP).
