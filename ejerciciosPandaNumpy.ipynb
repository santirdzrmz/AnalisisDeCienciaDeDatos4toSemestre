2.7

```python
def percentage(df):
  df_dead = df.loc[df["Survived"]==0]
  dead = len(df_dead)
  df_survived = df.loc[df["Survived"]==1]
  survivors = len(df_survived)
  total = dead+survivors
  print("Dead: ", dead/total*100)
  print("Survived: ", survivors/total*100)
df_upper = df.loc[df["Pclass"]==1]
print("First class")
percentage(df_upper)
df_middle = df.loc[df["Pclass"]==2]
print("Second class")
percentage(df_middle)
df_lower = df.loc[df["Pclass"]==3]
print("Third class")
percentage(df_lower)
```

```python
data = pd.read_csv('/content/adult.data.csv')
data.shape
```

4.1 How many men and women (*sex* feature) are represented in this dataset?

```python
male = 0
female = 0
i = 0
while i < len(data):
    if data.loc[i, "sex"] == "Male":
        male += 1
    else:
        female += 1
    i += 1
print("Number of males: " + str(male))
print("Number of females: " + str(female))
```

4.2 What is the average age (*age* feature) of women?

```python
women =  data.loc[data["sex"]=="Female"]
mean = women["age"].mean()
print("Average age of women: " + str(mean))
```

4.3 What is the percentage of German citizens (*native-country* feature)?

```python
total = len(data)
german = len(data.loc[data["native-country"]=="Germany"])
print("Percentage of German citizens: " + str(german/total*100))
```

4.4-4.5 What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year?

```python
earners =  data.loc[data["salary"]==">50K"]
losers =  data.loc[data["salary"]=="<=50K"]
print("Mean age of those who earn more than 50k:: " + str(earners["age"].mean()))
print("Standard deviation of age of those who earn more than 50k: " + str(earners["age"].std()))
print("Mean age of those who earn less than 50k: " + str(losers["age"].mean()))
print("Standard deviation of age of those who earn less than 50k: " + str(losers["age"].std()))
```

4.6 Is it true that people who earn more than 50K have at least high school education? (*education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)

```python
earners =  data.loc[data["salary"]==">50K"]
high_school_degrees = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
more_than_high_school = earners[earners['education'].isin(high_school_degrees)]
print("Percentage of those who earn more than 50k who have a high school degree or less: " + str(len(more_than_high_school)/len(earners)*100))
```

4.7 Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.

```python
sorted = data.groupby(["race", "sex"])["age"].describe()
print(sorted)
print("Maximum age of men of Amer-Indian-Eskimo race: " + str(sorted.loc["Amer-Indian-Eskimo", "Male"]["max"]))
```

4.8 What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?

```python
max = data["hours-per-week"].max()
print("Maximum number of hours a person works per week: " + str(max))

people = data.loc[data["hours-per-week"] == max]
print("Number of people who work " + str(max) + " hours per week: " + str(len(people)))

highest_earners = people.loc[people["salary"] == ">50K"]
print("Percentage of those who earn a lot (>50K) among them: " + str(len(highest_earners)/len(people)*100))
```

4.9 Find the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?

```python
sorted = data.groupby(["native-country", "salary"])["hours-per-week"].mean()
print(sorted)
print("Average time of work for those who earn a little for Japan: " + str(sorted.loc["Japan", "<=50K"]))
print("Average time of work for those who earn a lot for Japan: " + str(sorted.loc["Japan", ">50K"]))
```
