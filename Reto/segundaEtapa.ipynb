## Analisis exploratorio de Variables

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
y = datos['carencias_pob']

X = datos.drop(['carencias_pob', 'plp_pob'], axis=1)

#Variables escogidas de forma aleatoria
columnas = ['PSINDER', 'PAFIL_OTRAI', 'PRO_CRIEVA', 'VPH_EXCSA', 'VPH_CEL']

X = X[['PSINDER', 'PAFIL_OTRAI', 'PRO_CRIEVA', 'VPH_EXCSA', 'VPH_CEL']]
```

```python
X
```

```python
print(X.dtypes)

for col in X.columns:
    X[col] = pd.to_numeric(X[col], errors='coerce')

print(X.dtypes)
```

# Medidas de posicion no central

```python
print('Resumen estadístico:')
print(X.describe())
```

```python
num_cols = X.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(15, 5))
for i, col in enumerate(num_cols, 1):
    plt.subplot(1, len(num_cols), i)
    sns.boxplot(y=X[col])
    plt.title(f'Boxplot de {col}')
plt.tight_layout()
plt.show()
```

```python
def outliers(columna):
    Q1 = columna.quantile(0.25)
    Q3 = columna.quantile(0.75)
    IQR = Q3 - Q1
    return columna[(columna < Q1 - 1.5 * IQR) | (columna > Q3 + 1.5 * IQR)]
```

```python
for col in X.columns:
    outlier = outliers(X[col])
    print('Outliers in column ' + str(col) + ' of the dataframe X:')
    print(outlier)
```

# Analisis de distribucion de datos

```python
plt.figure(figsize=(15, 5))
for i, col in enumerate(num_cols, 1):
    plt.subplot(1, len(num_cols), i)
    sns.histplot(X[col], kde=True)
    plt.title(f'Histograma de {col}')
plt.tight_layout()
plt.show()
```

Todas las variables parecen estar sesgadas hacia la derecha y tienen forma asimetrica.

# Analisis de correlacion

```python
corr = X.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Mapa de calor de correlaciones entre variables")
plt.show()
```

Parece haber correlacion significativa entre VPH_EXCSA y VPH_CEL, esto puede deberse a colinealidad, variables que describen el mismo dato de diferente manera. Todas las variables parecen estar correlacionadas de manera significativa, con la mas baja siendo PRO_CRIEVA con PAFIL_OTRAI teniendo una correlacion de 0.71. Esto es malo, ya que ejemplifica que estas 5 variables no pueden tener todo el panorama para predecir la variable de interes.
